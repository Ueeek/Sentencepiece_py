```python
files = ! ls ./debug
```

```python
di="./debug/"
```

```python
pickles=[]
```

```python
for file in files:
    with open(di+file,"rb") as f:
        pickles.append(pickle.load(f))
```

```python
pickles[0]
```

```python
for pic in pickles:
    print("pic_gain=>",pic["gain"])
```

```python
for pic in pickles:
    if len(pic["pruned_voc"])==0:
        print("pruned_len=>",0)
    else:
        print("pruned_len=>",sum([len(v) for v in pic["pruned_voc"]])/len(pic["pruned_voc"]))
```
