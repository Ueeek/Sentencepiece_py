```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
! ls "../debug_no_approx"
```

```python
files=! ls | grep approx | grep -v ipy
```

```python
ps=[]
for file in files:
    with open(file,"rb") as f:
        ps.append((file,pickle.load(f)))
```

```python
for file,p in ps:
    print(file)
    X=[]
    Y=[]
    for key in p.keys():
        X.append(p[key]["approx"])
        Y.append(p[key]["no_approx"]*1000)
    Xnp=np.array(X)
    Ynp=np.array(Y)
    print(np.corrcoef(Xnp,Ynp))
    plt.scatter(Xnp,Ynp)
    plt.show()
        
```
