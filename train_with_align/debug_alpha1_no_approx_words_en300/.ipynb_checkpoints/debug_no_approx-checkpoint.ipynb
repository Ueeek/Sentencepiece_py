```python
prune_files = ! ls | grep prune | grep -v pruneloss
prune_losses = ! ls | grep pruneloss | grep -v diff
prune_loss_diffs = ! ls | grep pruneloss | grep diff
```

```python
prune_files
```

```python
prune_losses
```

```python
prune_loss_diffs
```

```python
prune_pics=[]
for file in prune_files:
    with open(file,"rb") as f:
        prune_pics.append(pickle.load(f))
```

```python
prune_losses_pics=[]
for file in prune_losses:
    with open(file,"rb") as f:
        prune_losses_pics.append((file,pickle.load(f)))
```

```python
prune_loss_diff_pics=[]
for file in prune_loss_diffs:
    with open(file,"rb") as f:
        prune_loss_diff_pics.append((file,pickle.load(f)))
```

```python
#
for p in prune_pics:
    print("bef:{} af:{}".format(p["obj_before"],p["obj_after"]))
```

```python
#
for file,p in prune_losses_pics:
    if "src" in file:
        continue
    print("remain")
    #print(p)
    for key,val in p["remain"].items():
        print("key:{} val:{}".format(key,val["AlignWords"]))
        pass
    print("remove")
    for key,val in p["remove"].items():
        print("key:{} val:{}".format(key,val["AlignWords"]))
        pass
```
